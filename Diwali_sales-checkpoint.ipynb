# Exploratory Data Analysis

**Gender**


```python
ax = sns.countplot(x='Gender',data=df,palette='Set2',hue='Gender',legend=False)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

ax = sns.barplot(x='Gender',y='Amount',data=sales,palette='Set2',hue='Gender',legend=False)
```

*From above graphs we can see that most of the buyers are female and even the purchasing power of females are higher.*

**Age**

```python
ax = sns.countplot(x='Age Group',data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales,hue='Age Group',legend=False)
```

*The Age Group of 26-35 has the highest numbers of female buyer*

**State**

```python
#Total no. of orders from top 10 states

orders = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
ax = sns.barplot(x='State',y='Orders',data=orders,hue='State',legend=False)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x='State',y='Amount',data=sales,palette='Set2',hue='State',legend=False)
```

*The above graph shows that Uttar Pradesh has the most number of buyers as well as has the highest purchasing power.*

**Marital Status**

```python
orders = df.groupby(['Marital_Status','Gender'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
ax = sns.barplot(x='Marital_Status',y='Orders',data=orders,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Marital_Status',y='Amount',data=sales,hue='Gender')
```

*In above graph we can see that the unmarried people are most of the buyers,but among the married couples females are the most buyers.*

**Occupation**

```python
sns.set(rc={'figure.figsize':(20,5)})

ax = sns.countplot(x='Occupation',data=df,palette='Set2',hue='Occupation',legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Occupation',y='Amount',data=sales,hue='Occupation',legend=False)
```

*From the graph it is clear that most of the buyers are from IT Sector , as well as it has the highest purchasing power.*

**Product Category**

```python
df.columns
```

```python
orders=df.groupby(['Product_Category'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.barplot(x='Product_Category',y='Orders',data=orders,hue='Product_Category',legend=False)
```

```python
sales = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.barplot(x='Product_Category',y='Amount',data=sales,hue='Product_Category',legend=False)
```

*From the above graphs we can see that most of the buyers belong to "Clothing and Apparel" category while the highest amount of the sold item belong to 
"Food" category.*

```python
order=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Orders',data=order,hue='Product_ID',legend=False)
```

```python
sales = df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.barplot(x='Product_ID',y='Amount',data=sales,hue='Product_ID',legend=False)
```

*The above graph shows that Product_ID P00265242 is the most sold product.*
